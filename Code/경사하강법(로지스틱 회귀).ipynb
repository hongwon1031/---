### 라이브러리 임포트

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 데이터 및 하이퍼 파라미터 설정

```python
x = np.array([1.0, 3.0, 5.0, 7.0])
y = np.array([0.0, 0.0, 1.0, 1.0])

w = 0               # 가중치
b = 0               # 절편
lrate = 0.1        # 학습률(learning rate)
epochs = 1000       # 에포크
m = float(len(x))   # 입력 데이터 개수
```

### 경사하강법 구현

```python
for i in range(epochs):
    z = w*x + b
    y_pred = 1 / (1 + np.exp(-z))
    dw = (1 / m) * sum((y_pred - y) * x)
    db = (1 / m) * sum((y_pred - y))
    w = w - lrate * dw
    b = b - lrate * db
print(f'w = {w}, b = {b}')
```

### 시각화

```python
z = w*x + b
y_pred = 1 / (1 + np.exp(-z))

plt.scatter(x, y)

plt.plot(x, np.zeros(len(z)) + 0.5, linestyle = 'dotted')
plt.plot(x, y_pred, color = 'red')
plt.show()
```
