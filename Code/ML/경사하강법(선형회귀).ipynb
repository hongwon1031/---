### 라이브러리 임포트

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 하이퍼 파라미터 설정

```python
x = np.array([0.0, 1.0, 2.0])
y = np.array([3.0, 3.5, 5.5])

w = 0               # 가중치
b = 0               # 절편
lrate = 0.01        # 학습률(learning rate)
epochs = 1000       # 에포크
m = float(len(x))   # 입력 데이터 개수
```

### 경사하강법

```python
for i in range(epochs):
    y_pred = w*x + b
    dw = (2/m) * sum(x * (y_pred - y))
    db = (2/m) * sum(y_pred - y)
    w = w - lrate * dw      # 가중치 수정
    b = b - lrate * db      # 절편 수정
print(f'w = {w}, b = {b}')
```

```python
y_pred = w*x + b # 예측값 생성
print(y_pred)
```

### 시각화

```python
plt.scatter(x, y)
plt.scatter(x, y_pred, color = 'red')
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color = 'red')
plt.show()
```
