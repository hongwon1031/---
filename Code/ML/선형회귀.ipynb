### 라이브러리 임포트

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

### 선형회귀 모델 생성

```python
reg = linear_model.LinearRegression()
```

### 데이터 생성

```python
x = [[0], [1], [2]] #2차원으로 생성해야함
y = [3, 3.5, 5.5]
```

### 학습 및 평가

```python
reg.fit(x,y)            # 모델 학습
score = reg.score(x,y)          # 모델 평가
y_pred = reg.predict(x) # 예측값 생성

print(score)
```

### 시각화

```python
plt.scatter(x, y, color = 'black')
plt.plot(x, y_pred, color = 'blue', linewidth = 3)
plt.show()
```
