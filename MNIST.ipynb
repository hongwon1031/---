### 라이브러리 불러오기

```python
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
```

### 데이터 확인

```python
digits = datasets.load_digits()
#plt.imshow(digits)
#plt.imshow(digits.data[0])
print(digits.keys())
print(digits.images[0])
#print(digits.data[0])
print(digits.data[0])
plt.imshow(digits.images[0],interpolation='nearest')
```

```python
n_sample = len(digits.images)
data = digits.images.reshape((n_sample,-1))
```

```python
print(len(data))
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,digits.target,test_size=0.2,random_state=42)
```

```python
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
scores = metrics.accuracy_score(y_test,y_pred)
print(scores)
```
